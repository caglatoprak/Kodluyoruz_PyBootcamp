```python
df_1=df.copy()
```

```python
df_selected = df_1.drop(['NationalITy', 'SectionID','PlaceofBirth', 'GradeID', "Topic", 'Semester', "Relation", 'ParentAnsweringSurvey', "ParentschoolSatisfaction", 'StudentAbsenceDays', "Class"], 
                        axis = 1)
```

```python
df_selected
```

```python
woman=df_selected[(df_selected["gender"]=='F')]
f=plt.figure(figsize=(10,5))
plt.bar(woman["StageID"],woman["raisedhands"],label="Woman")
plt.legend()
plt.title("raising hands according to the level of education in women")
plt.show()
```

```python
man=df_selected[(df_selected["gender"]=='M')]
f=plt.figure(figsize=(10,5))
plt.bar(man["StageID"],man["raisedhands"],label="Man")
plt.legend()
plt.title("raising hands according to the level of education in man")
plt.show()
```
