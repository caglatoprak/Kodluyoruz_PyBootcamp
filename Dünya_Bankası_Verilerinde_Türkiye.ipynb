# **Dünya Bankası Verilerinde Türkiye**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_excel("http://api.worldbank.org/v2/tr/country/TUR?downloadformat=excel",header=3)
```

```
data.head()
```

```
df=data.copy()
df.head()
```

```
df
```

İlk aşamada veri yapısını anlamaya çalıştım.

```
df.isnull().values.any()
```

```

df.info()
```

```
df.shape
```

```
df.columns
```

```
df_yeni=pd.DataFrame()
for col in df.iloc[:,4:]:
    df_baz = df.iloc[:,:4]
    df_baz["Year"]=col
    df_baz["Values"] = df[col]
    df_yeni = pd.concat([df_yeni,df_baz],ignore_index=True)
```

```
df_yeni.head()
```

```
turkiye=df_yeni.dropna()
```

```
turkiye.info()
```

```

turkiye["Year"] = turkiye["Year"].astype("int")
```

```
turkiye.head()
```

```
turkiye
```

```
#1960-2018 Arası Topla Nüfus, Yıllık Nüfus Artış Oranı,Doğurganlık oranı,Gençlerde doğurganlık oranı(15-19 yaş arası),Doğumda beklenen yaşam süresi(yıl),Ölüm Oranı(her 1000 canlı doğumu için)
nüfus=turkiye[(turkiye["Indicator Code"]=='SP.POP.TOTL')]

nüfusoran=turkiye[(turkiye["Indicator Code"]=='SP.POP.GROW')]

doğurganlıkoranı=turkiye[(turkiye["Indicator Code"]=='SP.DYN.TFRT.IN')]

genclerdedoğurganlıkoranı=turkiye[(turkiye["Indicator Code"]=='SP.ADO.TFRT')]

beklenenyasam=turkiye[(turkiye["Indicator Code"]=='SP.DYN.LE00.IN')]

ölümoranı=turkiye[(turkiye["Indicator Code"]=='SP.DYN.IMRT.IN')]


#f.savefig("grafik2.png")

fig = plt.figure(figsize=(20,12))

plt.subplot(3,3,1)   
plt.plot(nüfus["Year"],nüfus["Values"])
plt.title("Toplam Nüfus")
plt.tight_layout()

plt.subplot(3,3,2)
plt.plot(nüfusoran["Year"],nüfusoran["Values"])
plt.title("Yıllık Nüfus Artış Oranı(%)")
plt.tight_layout()

plt.subplot(3,3,3)
plt.plot(doğurganlıkoranı["Year"],doğurganlıkoranı["Values"])
plt.title("Doğurganlık oranı")
plt.tight_layout()

plt.subplot(3,3,4)
plt.plot(genclerdedoğurganlıkoranı["Year"],doğurganlıkoranı["Values"])
plt.title("Gençlerde doğurganlık oranı(15-19 yaş arası)")
plt.tight_layout()

plt.subplot(3,3,5)
plt.plot(beklenenyasam["Year"],beklenenyasam["Values"])
plt.title("Doğumda beklenen yaşam süresi(yıl)")
plt.tight_layout()

plt.subplot(3,3,6)
plt.plot(ölümoranı["Year"],ölümoranı["Values"])
plt.title("Ölüm Oranı(her 1000 canlı doğumu için)")
plt.tight_layout()


 


```

```
#1960-2018 Arası Kişi başına enerji kullanımı ve elektrik tüketimi
enerjikullanım=turkiye[(turkiye["Indicator Code"]=='EG.USE.PCAP.KG.OE')]
elektriktüketim=turkiye[(turkiye["Indicator Code"]=='EG.USE.ELEC.KH.PC')]



fig = plt.figure(figsize=(20,12))

plt.subplot(3,3,1)   
plt.plot(enerjikullanım["Year"],enerjikullanım["Values"],color="blue")
plt.title("Enerji Kullanımı(Kişi Başı Ham Petrol Kilogramına eş değer)")
plt.tight_layout()

plt.subplot(3,3,2)
plt.plot(elektriktüketim["Year"],elektriktüketim["Values"],color="red")
plt.title("Elektrik Tüketimi(kişi başı kWh)")
plt.tight_layout()
```

```
#2000 yılından itibaren İşsizlik Oranı (toplam işgücüne %'si)	
toplamissizlik=turkiye[(turkiye["Indicator Code"]=='SL.UEM.TOTL.ZS') & (turkiye.Year>2000)]

f=plt.figure(figsize=(12,6))
plt.bar(toplamissizlik["Year"],toplamissizlik["Values"],label="İşsizlik Oranı")
plt.legend()
plt.title("Yıllara Göre Toplam İşsizlik(toplam işgücüne %'si)")
plt.show()
```

```
#2000 yılından itibaren enflasyon ve toplam issizlik oranı
enflasyon=turkiye[(turkiye["Indicator Code"]=='FP.CPI.TOTL.ZG') & (turkiye.Year>2000)]
toplamissizlik=turkiye[(turkiye["Indicator Code"]=='SL.UEM.TOTL.ZS') & (turkiye.Year>2000)]

f=plt.figure(figsize=(12,6))

plt.bar(enflasyon["Year"],enflasyon["Values"],label="Enflasyon")

plt.bar(toplamissizlik["Year"],toplamissizlik["Values"],label="Toplam İşsizlik")

plt.legend()
#başlık koymak için
plt.title("Enflasyon, tüketici fiyatları (yıllık %)- ve İşsizlik Oranları, toplam (toplam işgücüne %'si)")

plt.show()
```

```
# 1960-2018 Arası Mal ve Hizmet İthalat ve İhtracatları(GSYİH %'si)
malvehizmetithalatları=turkiye[(turkiye["Indicator Code"]=='NE.IMP.GNFS.ZS')]
malvehizmetihracatları=turkiye[(turkiye["Indicator Code"]=='NE.EXP.GNFS.ZS')]


f=plt.figure(figsize=(12,6))


plt.plot(malvehizmetithalatları["Year"],malvehizmetithalatları["Values"],color="blue")

plt.plot(malvehizmetihracatları["Year"],malvehizmetihracatları["Values"],color="red")

plt.title("Mal ve Hizmet İthalat ve İhtracatları(GSYİH %'si)")
```

```
#2002-2018 Arası Yıllık Temiz Su Çekilmesi, CO2 Emisyonu(kişi başı metrik ton), Orman Alanı(km2)
yıllıktemizsucekilmesi=turkiye[(turkiye["Indicator Code"]=='ER.H2O.FWTL.ZS')& (turkiye.Year>2000)]
co2emisyonları=turkiye[(turkiye["Indicator Code"]=='EN.ATM.CO2E.PC')& (turkiye.Year>2000)]
ormanalanı=turkiye[(turkiye["Indicator Code"]=='AG.LND.FRST.K2')& (turkiye.Year>2000)]


fig = plt.figure(figsize=(20,12))

plt.subplot(3,3,1)   
plt.plot(yıllıktemizsucekilmesi["Year"],yıllıktemizsucekilmesi["Values"])
plt.title("Yıllık Temiz Su Çekilmesi")
plt.tight_layout()

plt.subplot(3,3,2)
plt.plot(co2emisyonları["Year"],co2emisyonları["Values"])
plt.title("CO2 Emisyonu(kişi başı metrik ton)")
plt.tight_layout()

plt.subplot(3,3,3)
plt.plot(ormanalanı["Year"],ormanalanı["Values"])
plt.title("Orman Alanı(km2)")
plt.tight_layout()
```
